### Общие слова

Python считается ООП-языком, хотя нито не заставляет на нем писать ООП-код. ООП-язык и ООП-парадигма это не синонимы.  
Он интерпретируемый и с динамической типизацией. Есть другие языки - похожие или противоположные по типу.  
Почему он популярен?

В этом уроке мы рассмотрим типы данных и основные операции без которых язык программирования не может существовать.

### Арифметика

```python
x + y # сложение
x - y # вычитание
# и так далее * / ** // *** /// %
```

````python
print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x ** y)
print(x // y)
print(x % y)
````

### Коллекции

Их больше чем я перечислил внизу, но в реальности можно ограничится даже меньшим числом коллекций

```python
# изменяемая, индексируемая #
[ True      , True        ] # список
{ 'm' : True, 'i' : False } # словарь - ключи только неизменяемые
( False     , True        ) # последовательность
{ True      , False       } # множество - только уникальные элементы
([False     , False      ]) # тоже множество - и здесь уникальные
```

````python
{(1, 2): '12'}
````

### Условия

В большинстве языков есть условный оператор, очевидно и тут тоже есть

```python
if True: 
    True
elif True:
    True
else: 
    True
```

```python
if True else False
```

```python
match True
    case True:
        True
    case _:
        True
```

### Циклы

Ну и не зря же мы коллекции выше перечислили

```python
while True:
    True
    break
    continue
```

```python
for i in range(4, 6):
    print(i)
else:
    print('')
```

```python
for i,j in enumerate('ij'):
    print(i)
```

```python
x = [x+1 for x in range(3)]
```
